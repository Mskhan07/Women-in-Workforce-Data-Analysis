```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
birth_rate_data = pd.read_csv("../data/crude-birth-rate.csv")
birth_rate_data.head()
```

```python
plt.figure(figsize=(20,10))
x_axis = birth_rate_data["DATE"]
y_axis = birth_rate_data["BIRTHS PER 1,000 PEOPLE"]
plt.xlabel("Years",fontsize=20, labelpad=10)
plt.ylabel("Birth Rate per 1,000 People",fontsize=20, labelpad=10)
plt.title ("Crude Birth Rate in US (1960 - 2017)",  fontsize=25, pad=15)
plt.plot(x_axis, y_axis, linewidth=2)
plt.savefig("../Images/crude-US-birth-rates.png", bbox_inches="tight")
plt.show()
```
