```python
# dependencies and setup
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read and display laborforce by race
labor_df = pd.read_csv('../data/laborforce-participation-rate.csv')
labor_df.head()
```

```python
# make data frames for women data
women_laborforce_df = pd.DataFrame({"Year": labor_df["Year"],
                                   "White Women": labor_df["White Women"],
                                   "Black Women": labor_df["Black Women"],
                                   "Asian Women": labor_df["Asian Women"],
                                   "Hispanic Women": labor_df["Hispanic Women"]})
white_women_df = pd.DataFrame({"Year": labor_df["Year"],
                                   "White Women": labor_df["White Women"]})
black_women_df = pd.DataFrame({"Year": labor_df["Year"],
                                   "Black Women": labor_df["Black Women"]})
asian_women_df = pd.DataFrame({"Year": labor_df["Year"],
                                   "Asian Women": labor_df["Asian Women"]})
hispanic_women_df = pd.DataFrame({"Year": labor_df["Year"],
                                   "Hispanic Women": labor_df["Hispanic Women"]})

# drop blank years
white_women = white_women_df.drop(white_women_df.index[:6])
black_women = black_women_df.drop(black_women_df.index[:24])
asian_women = asian_women_df.drop(asian_women_df.index[:52])
hispanic_women = hispanic_women_df.drop(hispanic_women_df.index[:25])

# make all x and y variables
x1 = white_women["Year"]
y1 = np.sort(white_women['White Women']).astype(float)
x2 = black_women["Year"]
y2 = np.sort(black_women['Black Women']).astype(float)
x3 = asian_women["Year"]
y3 = np.sort(asian_women['Asian Women']).astype(float)
x4 = hispanic_women["Year"]
y4 = np.sort(hispanic_women['Hispanic Women']).astype(float)
y1.min(),y2.min(),y3.min(),y4.min()
```

```python
# multiple line plot
plt.figure(figsize=(20,10))
plt.style.use('dark_background')
plt.title("Percentage of Women in Workforce by Race (1960 - 2016)", fontsize=25, pad=15)
plt.xlabel("Year",  fontsize=20, labelpad=10)
plt.ylabel("Percent", fontsize=20, labelpad=10)
plt.plot(x1,y1, color="red", linewidth=4, label="White Women")
plt.plot(x2,y2, color="lightgreen", linewidth=4, label="Black Women")
plt.plot(x3, y3, color="orange", linewidth=4, label="Asian Women")
plt.plot(x4, y4, color="skyblue", linewidth=4, label="Hispanic Women")
plt.legend(loc=2, fontsize=16)
plt.yticks(np.arange(30, 65, step=1), size='xx-large')
plt.xticks(size='xx-large')
plt.ylim(33,65)
plt.savefig("../Images/women-workforce-byrace-lineplot.png",bbox_inches="tight")
plt.show()
```
